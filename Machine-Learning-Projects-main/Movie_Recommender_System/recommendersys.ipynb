```python
move = move.merge(creds,on='title')
```

```python
move = move[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
move.head()
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
move.dropna(inplace=True)
```

```python
move['genres'] = move['genres'].apply(convert)
move.head()
```

```python
move['keywords'] = move['keywords'].apply(convert)
```

```python
move['cast'] = move['cast'].apply(convert)
move.head()
```

```python
move['cast'] = move['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
move['crew'] = move['crew'].apply(fetch_director)
```

```python
move.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
move['cast'] = move['cast'].apply(collapse)
move['crew'] = move['crew'].apply(collapse)
move['genres'] = move['genres'].apply(collapse)
move['keywords'] = move['keywords'].apply(collapse)
```

```python
move.head()
```

```python
move['overview'] = move['overview'].apply(lambda x:x.split())
```

```python
move['tags'] = move['overview'] + move['genres'] + move['keywords'] + move['cast'] + move['crew']
```

```python
new = move.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new['tags']).toarray()
vector.shape
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
                     
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new.iloc[i[0]].title)
```

```python
a=input("enter movie name for recommendations")
recommend(a)
```
